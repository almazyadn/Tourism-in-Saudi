####**(Primary Data) Survey** Cleaning and Preprocessing:
* Handled missing values:
categorical columns filled with "Unknown" - numerical columns filled with median values - remove completely empty rows.
* Standardized column names by converting to lowercase and replacing spaces with underscores.
* Ensured no repeated survey responses.

```python
# Cleaning Survey Data
survey_df.dropna(how='all', inplace=True)
survey_df.columns = [col.strip().replace(" ", "_").lower() for col in survey_df.columns]
categorical_cols = survey_df.select_dtypes(include=['object']).columns
numerical_cols = survey_df.select_dtypes(include=['number']).columns
survey_df[categorical_cols] = survey_df[categorical_cols].fillna("Unknown")
survey_df[numerical_cols] = survey_df[numerical_cols].apply(lambda x: x.fillna(x.median()))
survey_df.drop_duplicates(inplace=True)
# Save Cleaned Data
survey_df.to_csv("Cleaned_Survey.csv", index=False)
files.download("Cleaned_Survey.csv")
```

####**(Primary Data) X "twitter"** Cleaning and Preprocessing:
* Renamed columns for clarity (Date to timestamp, Tweet to content).
* Converted date format properly.
* Removed rows with empty content (tweets with no text).
* Cleaned tweet text (lowercase, removed URLs and links, and removed special characters).
* Remove Duplicate Tweets.

```python
# Cleaning X Data
twitter_df.rename(columns={"Date": "timestamp", "Tweet": "content"}, inplace=True)
twitter_df['timestamp'] = pd.to_datetime(twitter_df['timestamp'], errors='coerce')
twitter_df.dropna(subset=['content'], inplace=True)
twitter_df['content'] = twitter_df['content'].str.lower().str.replace(r'http\S+|www.\S+', '', regex=True)
twitter_df['content'] = twitter_df['content'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
twitter_df.drop_duplicates(inplace=True)
# Save Cleaned Data
twitter_df.to_csv("Cleaned_Twitter.csv", index=False)
files.download("Cleaned_Twitter.csv")
```

####**4.  Keyword Frequency Analysis**

```python
from collections import Counter
import re
from wordcloud import WordCloud

df_twitter["cleaned_content"] = df_twitter["content"].str.lower().apply(lambda x: re.sub(r'[^\w\s]', '', x))

all_words = " ".join(df_twitter["cleaned_content"]).split()
word_counts = Counter(all_words)

wordcloud = WordCloud(width=800, height=400, background_color="white", colormap="Blues").generate(" ".join(all_words))

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Most Frequent Terms in Tweets", fontsize=14)
plt.show()
```

####**5. Sentiment Analysis of Tweets**  


```python
from textblob import TextBlob

df_twitter["sentiment"] = df_twitter["cleaned_content"].apply(lambda x: TextBlob(x).sentiment.polarity)

df_twitter["sentiment_category"] = df_twitter["sentiment"].apply(lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Neutral"))

sentiment_counts = df_twitter["sentiment_category"].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
sentiment_counts.plot(kind='bar', color=["lightblue", "steelblue", "darkblue"], edgecolor="black", ax=ax)
ax.set_title("Sentiment Analysis of Tweets", fontsize=14)
ax.set_xlabel("Sentiment", fontsize=12)
ax.set_ylabel("Number of Tweets", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
```

Findings

The majority of tweets had a neutral sentiment, with a significant number of positive posts and very few negative ones. This suggests that most tourism-related discussions were either informational or promotional rather than complaints.

#### **Correlation Analysis**  


```python
import numpy as np

top_words = [word for word, count in word_counts.most_common(10)]

word_sentiment_df = pd.DataFrame(columns=["Word", "Positive", "Neutral", "Negative"])

for word in top_words:
    positive_count = df_twitter[df_twitter["sentiment_category"] == "Positive"]["cleaned_content"].str.contains(word).sum()
    neutral_count = df_twitter[df_twitter["sentiment_category"] == "Neutral"]["cleaned_content"].str.contains(word).sum()
    negative_count = df_twitter[df_twitter["sentiment_category"] == "Negative"]["cleaned_content"].str.contains(word).sum()
    word_sentiment_df = pd.concat([word_sentiment_df, pd.DataFrame({"Word": [word], "Positive": [positive_count], "Neutral": [neutral_count], "Negative": [negative_count]})], ignore_index=True)

word_sentiment_df.set_index("Word", inplace=True)
word_sentiment_df = word_sentiment_df.div(word_sentiment_df.sum(axis=1), axis=0).fillna(0)

plt.figure(figsize=(10, 6))
sns.heatmap(word_sentiment_df, cmap="Blues", annot=True, fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Keyword Frequency and Sentiment", fontsize=14)
plt.xlabel("Sentiment Category", fontsize=12)
plt.ylabel("Top Keywords", fontsize=12)
```

Findings

Words like “visit saudi” and “thislandiscalling” were associated with positive and neutral sentiment, indicating promotional content. Few strong negative correlations were found, suggesting that tourism-related keywords were not commonly linked to complaints.

```python
# Convert 'checkindate' to datetime format
tripadvisor_df_cleaned['checkindate'] = pd.to_datetime(tripadvisor_df_cleaned['checkindate'], errors='coerce')

# Extract the month from checkindate
tripadvisor_df_cleaned['month'] = tripadvisor_df_cleaned['checkindate'].dt.month

# Define seasons based on the month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the function to create a new column for season
tripadvisor_df_cleaned['season'] = tripadvisor_df_cleaned['month'].apply(get_season)

# Group by season and count the number of reviews (or check-ins)
seasonal_reviews = tripadvisor_df_cleaned.groupby('season')['numberofreviews'].sum()

# Plot the seasonal data
import matplotlib.pyplot as plt

# Bar plot for number of reviews per season
plt.figure(figsize=(10, 6))
seasonal_reviews.plot(kind='bar')
plt.title('Number of Reviews by Season (TripAdvisor)')
plt.xlabel('Season')
plt.ylabel('Number of Reviews')
plt.show()
```

##  Load Cleaned Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load cleaned datasets
base_path = "/content/drive"

datasets = {
    "Survey Data": pd.read_csv(base_path + "Cleaned_Survey.csv"),
    "Twitter Data": pd.read_csv(base_path + "Cleaned_Twitter.csv"),
    "TripAdvisor Data": pd.read_csv(base_path + "Cleaned_TripAdvisor.csv"),
    "Booking Data": pd.read_csv(base_path + "Cleaned_Booking.csv"),
    "Tourism Key Indicators": pd.read_csv(base_path + "Tourism_Key_Indicators.csv"),
    "Tourists by Province": pd.read_csv(base_path + "Tourists_Number_Provinces.csv"),
    "Tourists by Purpose": pd.read_csv(base_path + "Tourists_Number_Main_Purpose.csv"),
}
# Select relevant datasets
survey_data = datasets["Survey Data"]
tripadvisor_data = datasets["TripAdvisor Data"]
booking_data = datasets["Booking Data"]

# Filter only integer rating values from TripAdvisor
tripadvisor_data = tripadvisor_data[tripadvisor_data['rating'] % 1 == 0]

# Process Booking ratings by converting from 10-scale to 5-scale
booking_data['final_rating'] = np.round(booking_data['reviewscore'] / 2)

# Use final rating columns
tripadvisor_data['final_rating'] = tripadvisor_data['rating']

# Filtering and cleaning data
primary_column_name = "how_would_you_rate_your_overall_experience_in_saudi_arabia?"
secondary_column_name = "rating"

primary_data_filtered = survey_data[[primary_column_name]].dropna()
secondary_data_filtered = pd.concat([
    tripadvisor_data[['final_rating']].dropna(),
    booking_data[['final_rating']].dropna()
], axis=0).rename(columns={'final_rating': secondary_column_name})

# Convert Survey ratings to 1-5 scale
primary_data_filtered[primary_column_name] = primary_data_filtered[primary_column_name].apply(lambda x: np.clip(round(x), 1, 5))
secondary_data_filtered[secondary_column_name] = secondary_data_filtered[secondary_column_name].apply(lambda x: np.clip(round(x), 1, 5))

# Equalizing the sample sizes
sample_size = len(primary_data_filtered)
if len(secondary_data_filtered) >= sample_size:
    secondary_sampled = secondary_data_filtered.sample(n=sample_size, random_state=42)
else:
    print("Warning: Not enough data in secondary dataset after filtering. Using all available data.")
    secondary_sampled = secondary_data_filtered
```

The only common attribute between the primary and secondary data was the rating column. This section extracts the rating values, standardizes them to a 1-5 scale, and ensures consistency across sources. Survey ratings were rounded and clipped, while Booking.com ratings were converted from 10 to 5. To maintain fairness, we matched the sample size by randomly selecting a subset from the secondary data. If the secondary dataset was smaller, all available records were used. This process ensures a clean, standardized, and balanced dataset for accurate comparison.

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
survey_df = pd.read_csv("Cleaned_Survey.csv")
tripadvisor_df = pd.read_csv("Cleaned_TripAdvisor.csv")

# Clean and rename relevant columns
survey_df.columns = [col.strip().lower() for col in survey_df.columns]
tripadvisor_df.columns = [col.strip().lower() for col in tripadvisor_df.columns]

survey_df.rename(columns={
    '0-which_city_or_destination_in_saudi_arabia_did_you_visit_the_most?': 'city_visited',
    'what_factors_influenced_your_choice_of_destination?': 'factors_for_choice'
}, inplace=True)

tripadvisor_df.rename(columns={
    'addressobj/city': 'city',
    'rating': 'review_rating',
    'numberofreviews': 'review_count'
}, inplace=True)
```

```python
# Descriptive Summary
top_survey_cities = survey_df['city_visited'].value_counts().reset_index()
top_survey_cities.columns = ['city', 'num_visitors']

tripadvisor_summary = tripadvisor_df.groupby('city').agg({
    'review_rating': 'mean',
    'review_count': 'sum'
}).reset_index().rename(columns={
    'review_rating': 'avg_review_rating',
    'review_count': 'total_reviews'
})

# Combine survey and TripAdvisor data
combined_popularity = pd.merge(top_survey_cities, tripadvisor_summary, on='city', how='outer')
display(combined_popularity)
```

Counting how many visitors went to each city (from the survey), calculating average review ratings and total reviews from TripAdvisor, and merging both datasets to create a complete summary of destination popularity.



```python
#Clustering Prep
clustering_data = combined_popularity.dropna(subset=['avg_review_rating', 'total_reviews'])
X = clustering_data[['avg_review_rating', 'total_reviews']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Selecting the numeric features (average review rating and total reviews) and standardizing the values to prepare for clustering algorithms.



```python
#Elbow Method for K-Means
inertia = []
k_values = range(1, 10)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
sns.lineplot(x=k_values, y=inertia, marker='o')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Using the Elbow Method to determine the optimal number of clusters (k) for K-Means by plotting the inertia score. This helps identify how many distinct groups exist in the data.



```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clustering_data['kmeans_cluster'] = kmeans.fit_predict(X_scaled)

agglo = AgglomerativeClustering(n_clusters=3)
clustering_data['agglo_cluster'] = agglo.fit_predict(X_scaled)

# View final results
display(clustering_data)
```

Applying two clustering models — K-Means and Agglomerative Clustering — to group cities based on their popularity and review metrics. This helps identify patterns in tourist destinations and supports targeted tourism strategies.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
survey_df = pd.read_csv("Cleaned_Survey.csv")
survey_df.columns = [col.strip().lower() for col in survey_df.columns]

survey_df.rename(columns={
    'are_you_a_domestic_(saudi_resident)_or_international_tourist?': 'tourist_type',
    'what_was_your_estimated_total_expenditure_during_your_trip?': 'spending_range'
}, inplace=True)
```

```python
#Descriptive Analysis of Spending by Tourist Type
spending_data = survey_df[['tourist_type', 'spending_range']].dropna()

plt.figure(figsize=(12, 6))
sns.countplot(data=spending_data, x='spending_range', hue='tourist_type')
plt.title("Spending Ranges by Tourist Type")
plt.xlabel("Estimated Spending Range (SAR)")
plt.ylabel("Number of Tourists")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True, axis='y')
plt.show()
```

Domestic tourists tend to spend less, while international tourists are more common in higher spending ranges.


```python
#Encode Spending Ranges for Modelling
spending_order = {
    "1,000 to 5,000 Saudi Riyals": 1,
    "5,000 to 10,000 Saudi Riyals": 2,
    "10,000 to 15,000 Saudi Riyals": 3,
    "More than 15,000 Saudi Riyals": 4
}

model_data = survey_df[['tourist_type', 'spending_range']].dropna()
model_data = model_data[model_data['spending_range'].isin(spending_order.keys())]
model_data['spending_encoded'] = model_data['spending_range'].map(spending_order)
model_data['tourist_encoded'] = LabelEncoder().fit_transform(model_data['tourist_type'])
```

Data is now ready for classification — spending range as target, tourist type as feature.

```python
#Logistic Regression to Predict Spending Level
X = model_data[['tourist_encoded']]
y = model_data['spending_encoded']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

# Print classification results
report = classification_report(y_test, y_pred)
print(report)
```

 The model had some predictive power for mid-range spending, but accuracy was limited due to a small dataset.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
survey_df = pd.read_csv("Cleaned_Survey.csv")
survey_df.columns = [col.strip().lower() for col in survey_df.columns]
survey_df.rename(columns={
    'how_would_you_rate_your_overall_experience_in_saudi_arabia?': 'satisfaction',
    'are_you_a_domestic_(saudi_resident)_or_international_tourist?': 'tourist_type',
    'what_was_the_main_reason_for_visiting_saudi_arabia_in_that_season?': 'visit_reason',
    'how_many_days_did_you_spend_in_saudi_arabia?': 'duration',
    'what_was_your_estimated_total_expenditure_during_your_trip?': 'spending_range',
    'in_which_season_did_you_visit_saudi_arabia?': 'season'
}, inplace=True)
```

```python
q4_data = survey_df[['satisfaction', 'tourist_type', 'visit_reason', 'duration', 'spending_range', 'season']].dropna()
q4_data['satisfied'] = q4_data['satisfaction'].apply(lambda x: 1 if x >= 4 else 0)

features = ['tourist_type', 'visit_reason', 'duration', 'spending_range', 'season']
for col in features:
    q4_data[col] = LabelEncoder().fit_transform(q4_data[col])

X = q4_data[features]
y = q4_data['satisfied']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

Data is cleaned and satisfaction is encoded for classification.

```python
#Logistic Regression (Baseline)
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
y_pred = lr_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

Logistic Regression performed well overall, especially for predicting satisfied tourists.


```python
#Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
print(classification_report(y_test, rf_pred))
```

Random Forest had weaker accuracy and struggled to detect dissatisfied tourists.

```python
#Support Vector Machine (SVM)
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)
print(classification_report(y_test, svm_pred))
```

 SVM matched Logistic Regression in accuracy and performed best for satisfied tourists.
